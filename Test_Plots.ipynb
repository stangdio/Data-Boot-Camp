```python
import pandas as pd
import matplotlib.pyplot as mpl

df_crimes = pd.read_csv("Crimes_Dataset.csv")
df_suspects = pd.read_csv("Suspects_Dataset.csv")
```

```python
knife_crimes_by_monster_type = df_crimes[df_crimes['Crime Weapon'] == 'knife']
knife_crimes_by_monster_type = knife_crimes_by_monster_type.groupby("Monster involved")["Index_Crimes"].count()
knife_crimes_by_monster_type.sort_values(ascending=False)
knife_crimes_by_monster_type.plot(kind='bar')

# Thesis: Which Monster Tyypes commit the most knife crimes
```

```python
kidnapping_crimes_by_monster_type = (df_crimes[df_crimes["Crime Type"] == "kidnapping"].groupby("Monster involved").size())
kidnapping_crimes_by_monster_type.plot(kind="bar")

# Thesis: Which Monster Types committed the most kidnappings
```

```python
df_day_crimes = df_crimes[df_crimes['Time of Day'] == 'day']
kidnapping_crimes_by_monster_type = (df_day_crimes[df_day_crimes["Crime Type"] == "kidnapping"].groupby("Monster involved").size())
kidnapping_crimes_by_monster_type.plot(kind="bar")

# Which monster types committed the most kidnappings during the day
```

```python
df_hnliche_crimes = df_crimes[df_crimes['Crime Weapon'] == 'knife']
df_hnliche_crimes = df_hnliche_crimes[df_hnliche_crimes['Time of Day'] == 'day']
df_hnliche_crimes = df_hnliche_crimes[df_hnliche_crimes['Region']== 'village']
df_hnliche_crimes = df_hnliche_crimes[df_hnliche_crimes['Crime Type']== 'kidnapping']
df_hnliche_crimes = df_hnliche_crimes[df_hnliche_crimes['Evidence Found']== 'bones']
df_hnliche_crimes

# Which crimes in the past match the criteria of our crime (region = village, crime type = kidnapping, time of day = day, evidence found = bones)
```

```python
criminals_by_monster_type = (df_suspects[df_suspects["Criminal record"] == "Yes"].groupby("Monster").size())
criminals_by_monster_type.plot(kind="bar")
```

```python
crimes_by_monster_type = df_crimes.groupby("Monster involved")["Index_Crimes"].count()
crimes_by_monster_type.sort_values(ascending=False)
crimes_by_monster_type.plot(kind='bar')

# Thesis: Which Monster Tyypes commit the most crimes
```
