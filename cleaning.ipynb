```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Crimes_Dataset.csv')
df.info()
print(df.isnull().sum())
```

```python
# make everything lower case

def lowercase(df):
    object_cols = df.select_dtypes(include='object').columns 
    for attribute in object_cols:
        df[attribute] = df[attribute].str.lower()
    
    return df

lowercase(df)
```

```python
# change NaN to 'unknown'

def replaceNaN(df):
    df['Region'].fillna('unknown')
    df['Crime Weapon'].fillna('unknown')
    df['Crime Weapon'].replace('n/a', 'unknown')
    df['Evidence Found'].fillna('not found')

    return df

replaceNaN(df)
```
